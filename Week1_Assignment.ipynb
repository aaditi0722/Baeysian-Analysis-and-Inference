Q1. Suppose you have a jar with 4 jelly beans: 2 are strawberry-flavored, 1 is blueberry-flavored, and 1 is cinnamon-flavored. You draw one jelly bean at random from the jar.

    a. What is the sample space for this experiment?

    b. We define event A as the jelly bean drawn is strawberry-flavored and event B as The jelly bean drawn is not cinnamon-flavored. What are the probabilities of events A and B?

    c. Are events A and B mutually exclusive? Why or why not?


a. S= {s,s,b,c}
    s:strawberry flavoured
    b: blueberry-flavoured
    c: cinnamon-flavoured

b. Below code calculates the probabilities of events A and B

    

```python
def P(E,S):
  if set(E).issubset(set(S)) :
    return len(E)/len(S)
  else : return 0

A=['s','s']
S=['s','s','b','c']
B=['s','s','b']
print("Probability of Event A=",P(A,S))
print("Probability of Event A=",P(B,S))
```

c. These are not mutually exclusive events as the events A and B do have an intersection.


 Q2. Previously, we defined a **Python function P** to compute the probability of an event using the naive definition of probability. Generalize that function to compute the probability of events when they are
 not all equally likely. Use this new function to compute the probability of events A and B from the previous exercise.

 Hint: you can pass a third argument with the probability of each event.

```python


def P(E,S,prob):
  if set(E).issubset(set(S)):
    sum_prob=0
    for e in E:
      sum_prob+=prob[S.index(e)]
    return sum_prob
  else:
    return 0

A={'s'}
S=['s','s','b','c']
B={'s','b'}
prob=[0.5,0.5,0.25,0.25]

print("Probability of Event A=",P(A,S,prob))
print("Probability of Event B=",P(B,S,prob))
```

Q3.  Use PreliZ to explore different parameters for the Gaussian distributions. Use the methods `plot_pdf`, `plot_cdf`, and `plot_interactive`. We discussed the probability mass/density functions and the cumulative density function. But there are other ways to represent functions like the percentile point function ppf. Using the `plot_ppf` method of PreliZ, plot the percentile point function for the BetaBinomial and Gaussian distributions. Can you explain how the ppf is related to the cdf and pmf/pdf?

```python
import preliz
import matplotlib.pyplot as plt
_, ax = plt.subplots(3, 2, figsize=(12, 9), sharex="col")
preliz.BetaBinomial(alpha=10, beta=10, n=5).plot_ppf(ax=ax[0,0] ,legend="title")
preliz.Normal(0, 1).plot_ppf(ax= ax[0,1],legend="title")
preliz.BetaBinomial(alpha=10, beta=10, n=5).plot_pdf(ax=ax[1,0] ,legend="title")
preliz.Normal(0, 1).plot_pdf(ax= ax[1,1],legend="title")
preliz.BetaBinomial(alpha=10, beta=10, n=5).plot_cdf(ax=ax[2,0] ,legend="title")
preliz.Normal(0, 1).plot_cdf(ax= ax[2,1],legend="title")
```

From observing the above graphs, it is clear that Percentile Point function is the inverse of CDF or Cumulative Distribution Function. It basically maps its input probability p to a threshold value x so that the probability of random variable X being less or equal than x is p.



Q4. Use PreliZ to compute the moments for the SkewNormal distribution for a different combination of parameters. Generate random samples of different sizes,
like 10, 100, and 1,000, and see if you can recover the values of the first two moments (mean and variance) from the samples. Report your obervations.

```python


import preliz
import numpy as np

# Define the SkewNormal distribution with different parameters
sn1 = preliz.SkewNormal(10, 0.2, 1)
sn2 = preliz.SkewNormal(-5, 0.5, 0)
sn3 = preliz.SkewNormal(0, -0.5, 3)

# Compute the moments for each distribution
print("Moments for SkewNormal(10, 0.2, 1):")
print(sn1.moments(),"\n")
print("Moments for SkewNormal(-5, 0.5, 0):")
print(sn2.moments(),"\n")
print("Moments for SkewNormal(0, -0.5, 3):")
print(sn3.moments(),"\n")

# Generate random samples of different sizes
n1 = 10
n2 = 100
n3 = 1000

# Generate random samples from each distribution
samples_sn1_n1 = sn1.rvs(n1)
samples_sn1_n2 = sn1.rvs(n2)
samples_sn1_n3 = sn1.rvs(n3)
samples_sn2_n1 = sn2.rvs(n1)
samples_sn2_n2 = sn2.rvs(n2)
samples_sn2_n3 = sn2.rvs(n3)
samples_sn3_n1 = sn3.rvs(n1)
samples_sn3_n2 = sn3.rvs(n2)
samples_sn3_n3 = sn3.rvs(n3)
# Calculate the sample mean and variance for each sample
sample1_mean1 = np.mean(samples_sn1_n1)
sample1_var1 = np.var(samples_sn1_n1)

sample1_mean2 = np.mean(samples_sn1_n2)
sample1_var2 = np.var(samples_sn1_n2)

sample1_mean3 = np.mean(samples_sn1_n3)
sample1_var3 = np.var(samples_sn1_n3)

sample2_mean1 = np.mean(samples_sn2_n1)
sample2_var1 = np.var(samples_sn2_n1)

sample2_mean2 = np.mean(samples_sn2_n2)
sample2_var2 = np.var(samples_sn2_n2)

sample2_mean3 = np.mean(samples_sn2_n3)
sample2_var3 = np.var(samples_sn2_n3)

sample3_mean1 = np.mean(samples_sn3_n1)
sample3_var1 = np.var(samples_sn3_n1)

sample3_mean2 = np.mean(samples_sn3_n2)
sample3_var2 = np.var(samples_sn3_n2)

sample3_mean3 = np.mean(samples_sn3_n3)
sample3_var3 = np.var(samples_sn3_n3)


# Print the sample moments
print("Sample moments for SkewNormal(10, 0.2, 1) with n = 10:")
print("Mean:", sample1_mean1)
print("Variance:", sample1_var1,"\n")

print("Sample moments for SkewNormal(10, 0.2, 1) with n = 100:")
print("Mean:", sample1_mean2)
print("Variance:", sample1_var2,"\n")

print("Sample moments for SkewNormal(10, 0.2, 1) with n = 1000:")
print("Mean:", sample1_mean3)
print("Variance:", sample1_var3,"\n")

print("Sample moments for SkewNormal(-5, 0.5, 0) with n = 10:")
print("Mean:", sample2_mean1)
print("Variance:", sample2_var1,"\n")

print("Sample moments for SkewNormal(-5, 0.5, 0) with n = 100:")
print("Mean:", sample2_mean2)
print("Variance:", sample2_var2,"\n")

print("Sample moments for SkewNormal(-5, 0.5, 0) with n = 1000:")
print("Mean:", sample2_mean3)
print("Variance:", sample2_var3,"\n")

print("Sample moments for SkewNormal(0, -0.5, 3) with n = 10:")
print("Mean:", sample3_mean1)
print("Variance:", sample3_var1, "\n")

print("Sample moments for SkewNormal(0, -0.5, 3) with n = 100:")
print("Mean:", sample3_mean2)
print("Variance:", sample3_var2, "\n")

print("Sample moments for SkewNormal(0, -0.5, 3) with n = 1000:")
print("Mean:", sample3_mean3)
print("Variance:", sample3_var3, "\n")


```

From observations, it is clear that as the value of n (number of samples)  increases, the values of moments get closer and closer to the actual values.

Also, the value of alpha determines the deviation of the mean and variance of the distribution from the mean and variance parameters used to make the skewnormal distributions.
For instance, in the second distribution, alpha=0, hence, the value of the moments (mean and variance) is exactly equal to the value of parameters used to make this distribution.

Hence, alpha represents the deviation of the skewnormal from a normal distribution.
